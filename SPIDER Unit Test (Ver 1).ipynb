```python
weak = SPIDER(kind='weak')

X_w, y_w = weak.fit_resample(X, y)

df_w = pd.DataFrame({'x': X_w[:, 0], 'y': X_w[:, 1], 'class': y_w})
df_w.sort_values('x')
```
